```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import IsolationForest
```

```python
from google.colab import files
uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])
df.head()
```

```python
df.info()
df.describe()
```

```python
df['Class'].value_counts()
```

```python
# Check missing values
df.isnull().sum()
```

```python
X = df.drop('Class', axis=1)
y = df['Class']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_test)
```

```python
print("Logistic Regression Report")
print(classification_report(y_test, y_pred_lr))
```

```python
dt = DecisionTreeClassifier(max_depth=5, random_state=42)
dt.fit(X_train, y_train)

y_pred_dt = dt.predict(X_test)
```

```python
print("Decision Tree Report")
print(classification_report(y_test, y_pred_dt))
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, y_pred_lr), annot=True, fmt='d')
plt.title("Confusion Matrix - Logistic Regression")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
iso = IsolationForest(contamination=0.01, random_state=42)
y_pred_iso = iso.fit_predict(X_scaled)

# Convert output: -1 → Fraud, 1 → Normal
y_pred_iso = np.where(y_pred_iso == -1, 1, 0)
```

```python
print("Isolation Forest Report")
print(classification_report(y, y_pred_iso))
```

```python
def predict_transaction(transaction):
    transaction_scaled = scaler.transform([transaction])
    prediction = lr.predict(transaction_scaled)
    return "FraUD" if prediction[0] == 1 else "Legitimate"

# Example
predict_transaction(X.iloc[0])
```
